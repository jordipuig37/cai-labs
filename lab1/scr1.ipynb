```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(np.log(apellidos["Orden"]), np.log(apellidos["Apellido 1"]))
```

```python
def abline(slope, intercept):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '-')
```

```python
%matplotlib notebook
plt.plot(np.log(apellidos["Orden"]), np.log(apellidos["Apellido 1"]))
plt.xlabel("log(orden)")
plt.ylabel("log(apellido1)")
abline(slope, intercept)
plt.show()
```
