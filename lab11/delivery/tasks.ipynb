```python
from igraph import Graph
from igraph import plot
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
```

## First task

```python
p_vals = np.concatenate([[0], np.logspace(-4, 0)])
```

```python
avg_len = []
clust_coef = []
val1 = 0
val2 = 0
N = 1000 #Â we can vary N
for p in p_vals:
    watts = Graph.Watts_Strogatz(1, N, 2, p)
    if p == 0:
        val1 = watts.average_path_length()
        val2 = watts.transitivity_undirected()
    avg_len.append(watts.average_path_length() / val1)
    clust_coef.append(watts.transitivity_undirected() / val2)
```

```python
fig, ax = plt.subplots()

ax.plot(p_vals, clust_coef)
ax.plot(p_vals, avg_len)
ax.set(xscale="log")
ax.set(title = "", xlabel = "p", ylabel = "Cluster coeff. and Avg. Shortest Path")
ax.legend(["Cluster Coeff.", "Avg. Shortest Path"],loc = "best")
fig.savefig("task1Python.png")
```
